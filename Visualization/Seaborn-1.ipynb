**1.Scatter Plot**

A scatter plot is a type of plot used to visualize the relationship between two numerical variables and is useful understanding trends and correlations variables.

```python
#import

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# load dataset 
sns.set(style="darkgrid")
tips = sns.load_dataset("tips") 
tips.head()
```

```python
# 12 different scatter plots of tips dataset

#subplots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 20))

# 1D list for indexing
axes = axes.flatten()

#scatter plots
plots = [
    # sns.scatterplot(x="",y=,hue="",size="")
    {"x": "total_bill", "y": "tip"},
    {"x": "total_bill", "y": "tip", "hue": "smoker"},
    {"x": "total_bill", "y": "tip", "hue": "smoker", "style": "time"},
    # lunch' style is dot,dinner is X
    {"x": "total_bill", "y": "tip", "hue": "size", "palette": "ch:r=-.5,l=.75"},
    {"x": "total_bill", "y": "tip", "size": "size"},
    {"x": "total_bill", "y": "tip", "size": "size", "sizes": (15, 200)},
    {"x": "total_bill", "y": "tip"},
    {"x": "total_bill", "y": "tip", "hue": "time"},
    {"x": "total_bill", "y": "tip", "hue": "smoker", "style": "time"},
    {"x": "total_bill", "y": "tip", "hue": "size", "palette": "ch:r=-.5,l=.75"},
    {"x": "total_bill", "y": "tip", "size": "size"},
    {"x": "total_bill", "y": "tip", "size": "size", "hue": "smoker"},
]

for i, params in enumerate(plots):
    sns.scatterplot(data=tips, ax=axes[i], **params)
    axes[i].set_title(f"Scatter Plot {i+1}")

plt.tight_layout()
plt.show()
```

**3.Line Plot**

```python
# load dataset 
sns.set(style="darkgrid")
fmri = sns.load_dataset("fmri") 
fmri.head()
```

```python


#subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 20))

# 1D list for indexing
axes = axes.flatten()

plots = [
    {"x": "timepoint", "y": "signal"},
    {"x": "timepoint", "y": "signal","hue":"event"},
    {"x": "timepoint", "y": "signal","size":"region"},
    {"x": "timepoint", "y": "signal","hue":"region","style":"event"},
    {"x": "timepoint", "y": "signal","hue":"region","palette": "ch:r=-.5,l=.75"},
    {"x": "timepoint", "y": "signal","hue":"event","style":"event","markers":True,"dashes":False},
    {"x": "timepoint", "y": "signal","hue":"subject"},
    {"x": "timepoint", "y": "signal","hue":"subject","sizes":(.25, 2.5)}]

for i, params in enumerate(plots):
    sns.lineplot(data=fmri, ax=axes[i], **params)
    axes[i].set_title(f"Line Plot {i+1}")

plt.tight_layout()
plt.show()
```

**4.Boxplot**

Summarize distribution of a numeric variable across categories.

```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```

Use for medians, quartiles, and outliers per category.

**5.Violinplot**

Combination of boxplot + distribution shape.

```python
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.show()
```

**6.Barplot**

Compare averages (or other estimates) across categories.

```python
sns.barplot(data=tips, x="day", y="tip")
plt.show()
```

**7.Countplot**

Show frequency counts for categorical variables.

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.show()
```

**8.Pairplot**

Show pairwise relationships among multiple variables.

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

**9.Heatmap**

Visualize correlations or matrix-like data.

```python
numeric_tips = tips.select_dtypes(include="number")
sns.heatmap(numeric_tips.corr(), annot=True, cmap="coolwarm")
plt.show()
```
